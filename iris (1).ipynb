```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris.species.unique()
```

```python
sns.kdeplot(iris.loc[(iris['species']=='setosa'),'sepal_length'],color='g',label='setosa')
plt.legend()
```

```python
sns.kdeplot(iris.loc[(iris['species']=='setosa'),'sepal_length'],color='g',label='setosa')
sns.kdeplot(iris.loc[(iris['species']=='versicolor'),'sepal_length'],color='r',label='versicolor')
plt.legend()
```

```python
sns.kdeplot(iris.loc[(iris['species']=='setosa'),'sepal_length'],color='g',label='setosa')
sns.kdeplot(iris.loc[(iris['species']=='versicolor'),'sepal_length'],color='r',label='versicolor')
sns.kdeplot(iris.loc[(iris['species']=='virginica'),'sepal_length'],color='b',label='verginica')
#sns.add_legend()
plt.legend()
```

```python
sns.relplot(data=iris,x="sepal_length",y="sepal_width",color='b')
```

```python
sns.relplot(data=iris,x="petal_length",y="petal_width",color='b')
```

```python
labels='Car','Truck','Bus','Train'
sizes=[250,350,400,500]
colors=['b','r','m','g']
explode=(0.1,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
